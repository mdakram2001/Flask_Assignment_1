```python
"""Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans:

Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular
tools or libraries. It is designed to be simple and lightweight, providing a flexible and easy-to-use framework for building
web applications

Advantages of Flask Framework:
1.Lightweight and Minimalistic
2.Easy to Learn and Use
3.Flexibility
4.Modular and Extensible"""
```

```python
"""Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
Ans:"""

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
"""Q3. What is App routing in Flask? Why do we use app routes?
Ans:

In Flask, app routing refers to the process of mapping URLs (routes) to specific functions or view handlers within 
the Flask application.The app.route() decorator in Flask is used to define routes. It allows you to specify the URL 
pattern that triggers a specific function when accessed.
We use app routes in Flask to define the different URLs or routes that our web application will respond to. 
Each route is associated with a specific function or view that will be executed when that route is accessed, 
allowing us to create different pages or endpoints for our application."""
```

```python
"""Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Ans:"""

from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Company Name: ABC Corporation<br> Location: India<br> Contact Detail: 999-999-9999</h1>"

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
"""Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans:

In Flask, the url_for() function is used for URL building. This function generates a URL for a given endpoint and any
additional parameters provided"""

from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Welcome to Home Page</h1>"

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to Our Company</h1>"

@app.route("/welcome/<cname>")
def cwelcome(cname):
    return f"<h1>Welcome to {cname}</h1>"

@app.route("/<name>")
def company(name):
    return f"<h1>Company Name is {name}</h1>"

@app.route("/loc/<location>")
def companyloc(location):
    return f"<h1>Company is located in {location}</h1>"

@app.route("/<int:contact>")
def companycont(contact):
    return f"<h1>Contact Us at #{contact}</h1>"

if __name__=="__main__":
    #app.run(host="0.0.0.0")
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('welcome'))
        print(url_for('cwelcome', cname='ABS International'))
        print(url_for('company', name='ABS International'))
        print(url_for('companyloc', location='India'))
        print(url_for('companycont', contact=9954118971))
```
